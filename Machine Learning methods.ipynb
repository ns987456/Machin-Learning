```python
x_train, x_test, y_train, y_test = train_test_split(x,y)

x_train_M=x_train
x_test_M=x_test

score_svm=[]
for i in range(len(features_corr)-1):
    x_train_M.drop(features_corr.index[i], inplace=True, axis=1)
    x_test_M.drop(features_corr.index[i], inplace=True, axis=1)
    C_range = np.logspace(-2, 10, 13)
    gamma_range = np.logspace(-9, 3, 13)
    param_grid = dict(gamma=gamma_range, C=C_range)
    cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
    grid = GridSearchCV(SVC(), param_grid=param_grid, cv=cv)
    model=grid.fit(x_train_M, y_train)
    y_predict_svm = model.predict(x_test_M)
    score_svm.append( accuracy_score(y_test, y_predict_svm))
 
```

```python
names=[]
for i in range(len(features_corr)-1):
    names.append("M"+str(i))

models=pd.DataFrame({'model' :names,'accuracy' :score_svm})
models
```

```python
import matplotlib.pyplot as plt

plt.plot(models.model,models.accuracy)
plt.title('title name')
plt.xlabel('models')
plt.ylabel('accuracy')
plt.show()
```
